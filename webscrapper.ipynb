```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
dfs = pd.read_excel('project 1.xlsx', sheet_name=None)
df=dfs['textile']
```

```python
df.head()
```

```python
headlines_by_company = {}

for index, row in df.iterrows():
    company = row['Company Name']
    url = row['news']
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    all_headlines = soup.findAll("a", {"class": "DY5T1d RZIKme"})
    all_dates = soup.findAll("div",{"SVJrMe"})

    if company not in headlines_by_company:
        headlines_by_company[company] = {}

    for i in range(len(all_headlines)):
        headline = all_headlines[i].text
        date = all_dates[i].text
        if not date:
            date = 'N/A'

        if headline not in headlines_by_company[company]:
            headlines_by_company[company][headline] = date

df_headlines = pd.DataFrame(columns=['Company Name', 'Headlines', 'Dates'])

for company, headlines_dict in headlines_by_company.items():
    for headline, date in headlines_dict.items():
        df_headlines = df_headlines.append({
            'Company Name': company,
            'Headlines': headline,
            'Dates': date
        }, ignore_index=True)

print(df_headlines.head())
```
